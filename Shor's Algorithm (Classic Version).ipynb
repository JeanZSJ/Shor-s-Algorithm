```python
## Classical Pre-Processing

## Problem: Given N, where N is an odd number, and N is not an odd power, factor N into 2 integers p & q > 1

## Let us pick N = 15

import numpy as np
import random as rd
import math
```

```python
## I am going to pick two large primes and multiply them together: 197, 199
p1 = 197
p2 = 199

N = p1* p2

# pick a random number a between 2 and N-1, inclusive
a = rd.randint(2, N-1)
print(a)
```

```python
## this is the euclidean algorithm for finding greatest common factor


def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print("The gcd is:")
gcd(N,a)


# If gcd(N,a) neq 1, we have found a nontrivial factor of N. We are done.
# Else, we know that a and N are coprime, and we can proceed with the algorithm.
## for example, 14 = 2*7; 15= 3*5; the greatest common factor they share is 1
```

```python
## Now we want to find the smallest order r such that a^r mod N = 1
# We can implemenent this classically:

## this version of findOrder is inefficient in memory

# I first tried to implement findOrder with recurison, but it turns out to be very memory intensive.
# Now I use a while loop instead.

#def findOrder(num_a, num_b, order):
#    if num_a**order % num_b == 1:
#        return order
#    else:
#        return findOrder(num_a, num_b, order + 1)



def findOrder(a, N):
    
    if gcd(a,N) != 1:
        return N
        
    order = 1
    value = a % N
    
    while value != 1:
        value = (value * a) % N
        order += 1

    return order
    
## This version is much better!

order = findOrder(a, N)
print("The order is:")
print(order)
```

```python
## if our order is odd, we cannot proceed with the current algorithm. we have to go back & look for another a

## However, if our order is even, we can proceed.

if order % 2 ==0:
    print("Yes, we can proceed with this order.")
else:
    print("No, the order is odd, so we have to go back and find a new a")

new_order = order // 2
## // 2 is doing integer division by 2
print(new_order)


## now we have (a^(r/2)-1)(a^(r/2)+1) = 0 mod N; which means that (a^(r/2)-1)(a^(r/2)+1) divides N
## if we compute a**new_order, the number is too large  & would likely exceed the limit for integer string conversion

#split_1 = a**new_order - 1
#split_2 = a**new_order + 1

## instead, to simplify the arithmetic, we can utilize the property that
# x= a^(r//2) mod N; 
# then (a^(r/2)-1)(a^(r/2)+1) = 0 mod N = (x - 1) (x + 1)

x = pow(a, new_order, N)
# this means compute a^new_order mod N
print("a^(r//2) is equivalent to")
print(x)

split_1 = x - 1
split_2 = x + 1

print("The smaller factor is:")
print(split_1)
print("The bigger factor is:")
print(split_2)
```

```python
if split_2 % N == 0:
    print("Go back & find another a")
else:
    print("Yes, we can proceed")
```

```python
p = gcd(split_2, N)
q = int(N/p)

if p==1:
    print("we can't obtain useful information about the prime factors of N. Go back and choose another a")
else:
    print("A factor of N is")
    print(p)
    print("The other factor of N is")
    print(q)
    
```

```python
# An example of RSA encryption

# Choose 2 primes:
p = 61
q = 53
N = p*q
print(N)

# Compute Euler's totient
phi = (p-1)*(q-1)
print(phi)

# Choose public exponent e
## to do that, you want to pick something coprime with 3120
e = 17

# Find private key d
## we want e*d mod 3120 = 1
d = pow(e, -1, phi)
# this is solvable using the extended euclidean algorithm
print(d)

# the reason we want to do that is that the sender encryps message with public key

# c = m**e mod N

# you want to use your private key to recover the original message

# m = c**d mod N

# this works because: m**(ed) = m mod N

# Now, we have our public key N = 3233, e = 17
# Our private key d = 2753

# Now we encrypt our message:
m = 65
c = pow(m, e, N)
print("the ciphertext is:")
print(c)

#now we decrypt the message
decrypted_message = pow(c, d, N)
print("the decrypted message is:")
print(decrypted_message)

#if a hacker wants to access the message, they have to factor N = p*q
### Perhaps with the classical Shor's algorithm

# The power of RSA encryption lies in the fact that: the factoring step is very hard!

# The hacker would have to factor 3233 first! Which is exponentially hard for classical comptuers
# 3233 = 61 * 51

# Then they have to compute euler's totient
# phi = (61-1)*(51-1)

# Then they have to recover d, 17d mod 3120 = 1 using pow(17, -1, 3120)
# Now they have the private key d, the hacker cna access the message.
```
